```python
# фильтрация пользователей и книг по активности
count_users = ratings['User-ID'].value_counts()
active_users = count_users[count_users >= 50].index

count_books = ratings['ISBN'].value_counts()
popular_books = count_books[count_books >= 50].index

ratings_filt = ratings[
(ratings['User-ID'].isin(active_users)) & (ratings['ISBN'].isin(popular_books))
]
print(f'Количество оставшихся оценок = {len(ratings_filt)}')
```

```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# подготовка данных
data = ratings_filt[['User-ID', 'ISBN', 'Book-Rating']].copy()
data.columns = ['user_id', 'isbn', 'rating']
reader = Reader(rating_scale=(0, 10))
dataset = Dataset.load_from_df(data[['user_id', 'isbn', 'rating']], reader)
trainset, testset = train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
# оптимизация гиперпараметров
from surprise.model_selection import GridSearchCV

param_grid = {
    'n_factors': [20, 50, 70, 100],
    'n_epochs': [10, 20, 30],
    'lr_all': [0.002, 0.005, 0.01],
    'reg_all': [0.02, 0.1]
}
gs = GridSearchCV(SVD, param_grid, measures=['rmse'], cv=3)
gs.fit(dataset)
print(gs.best_params['rmse'])
```

```python
model = SVD(n_factors=20, n_epochs=10, lr_all=0.005, reg_all=0.1, random_state=42)
model.fit(trainset)
predictions = model.test(testset)
```

```python
#получение рекомендаций
def get_recommendation(model, user_id, isbn_title, n=10):
    #все книги
    all_isbns = trainset.all_items()
    all_isbns_raw = [trainset.to_raw_iid(iid) for iid in all_isbns]
    # книги, которые уже оценены пользователем
    user_ratings = data[data['user_id'] == user_id]
    rated_isbns = set(user_ratings['isbn'].values)
    candidates = [isbn for isbn in all_isbns_raw if isbn not in rated_isbns]
    # рекомендации новых книг для пользователя
    predictions = []
    for isbn in candidates:
        try:
            pred = model.predict(user_id, isbn)
            predictions.append((isbn, pred.est))
        except:
            continue
    
    # isbn в названия
    predictions.sort(key=lambda x: x[1], reverse=True)
    top_n = predictions[:n]
    recommendations = []
    for isbn, est_rating in top_n:
        title = isbn_title.get(isbn, 'Unknown Title')
        recommendations.append((title, round(est_rating, 2)))
        
    return recommendations
```

```python
# вывод рекомендаций для случайного пользователя
import random

s_user_inner_id = random.choice(trainset.all_users())
s_user_id = trainset.to_raw_uid(s_user_inner_id)

print(f'Рекомендации книг для пользователя ID: {s_user_id}')
recommendations = get_recommendation(model, s_user_id, isbn_title, n=5)

for i, (title, rating) in enumerate(recommendations, 1):
    print(f'Книга "{title}" - рейтинг {rating}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

isbn_to_author = dict(zip(books_filt['ISBN'], books_filt['Book-Author']))

all_authors = list(set(isbn_to_author.values()))
author_to_idx = {author: idx for idx, author in enumerate(all_authors)}

def author_vector(isbn):
    author = isbn_to_author.get(isbn, 'Unknown')
    vector = np.zeros(len(all_authors))
    if author in author_to_idx:
        vector[author_to_idx[author]] = 1
    return vector

def diversity_k(model, user_id, isbn_to_author, k=10):
    recommendations = get_recommendation(model, user_id, isbn_title, n=k)
    isbns = [
        books_filt[books_filt['Book-Title'] == title]['ISBN'].values[0]
        for title, _ in recommendations if len(books_filt[books_filt['Book-Title'] == title]) > 0
    ]
    if len(isbns) < 2:
        return 1.0
    vectors = np.array([author_vector(isbn) for isbn in isbns])
    sim_matrix = cosine_similarity(vectors)
    triu_indices = np.triu_indices_from(sim_matrix, k=1)
    avg_similarity = np.mean(sim_matrix[triu_indices])
    return 1 - avg_similarity

sample_user_id = trainset.to_raw_uid(trainset.all_users()[0])
diversity_score = diversity_k(model, sample_user_id, isbn_to_author, k=10)
print(f'Diversity (ID = {sample_user_id}): {diversity_score:.4f}')
```

```python
from collections import defaultdict

# precision & recall
def precision_recall_k(predictions, k=10, threshold=8.0):
    user_est_true = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        user_est_true[uid].append((est, true_r))
    
    precisions = {}
    recalls = {}
    for uid, user_ratings in user_est_true.items():
        user_ratings.sort(key=lambda x: x[0], reverse=True)
        top_k = user_ratings[:k]
        n_rel_and_rec = sum((true_r >= threshold) for (_, true_r) in top_k)
        n_rec = len(top_k)
        n_rel = sum((true_r >= threshold) for (_, true_r) in user_ratings)
        precisions[uid] = n_rel_and_rec / n_rec if n_rec > 0 else 0
        recalls[uid] = n_rel_and_rec / n_rel if n_rel > 0 else 0
    return precisions, recalls

precisions, recalls = precision_recall_k(predictions, k=10, threshold=8.0)

avg_precision = sum(precisions.values()) / len(precisions) if precisions else 0
avg_recall = sum(recalls.values()) / len(recalls) if recalls else 0

print(f'Precision = {avg_precision:.4f}')
print(f'Recall = {avg_recall:.4f}')

```
